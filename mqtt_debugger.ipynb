```python
import os

# roda mosquitto via docker
os.system("""
docker run -d \
  --name mqtt-broker \
  -p 1883:1883 \
  eclipse-mosquitto
""")

print("✅ Broker MQTT iniciado na porta 1883")
```

```python
import paho.mqtt.client as mqtt
from datetime import datetime

BROKER = "localhost"
PORT = 1883

def on_connect(client, userdata, flags, rc):
    print("✅ Conectado ao broker! Código:", rc)
    client.subscribe("#")
    print("📡 Escutando TODOS os tópicos...\n")

def on_message(client, userdata, msg):
    try:
        payload = msg.payload.decode()
    except:
        payload = msg.payload

    print("="*50)
    print("🕒", datetime.now())
    print("📍 Tópico:", msg.topic)
    print("📦 Payload:", payload)

client = mqtt.Client()

client.on_connect = on_connect
client.on_message = on_message

client.connect(BROKER, PORT, 60)
client.loop_start()
```
